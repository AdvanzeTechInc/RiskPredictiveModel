```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import re
```

```python
df=pd.read_excel("/content/Claims data_1.xlsx")
df
```

```python
#To keep only classcode number
df["ClassCode"] = df["ClassCode"].str.replace("ClassCode:", "")
df
```

```python
#to take the complete zipcode value
df["ZipCode"] = df["ZipCode"].str.replace("-", "")
df
```

```python
#rows where the 'Amount' column contains valid numeric values & remove any rows with non-numeric or missing 'Amount' values
df = df[pd.to_numeric(df['Amount'], errors='coerce').notna()]
df
```

```python
#search for Amount value if there is a decimal point in it and then keep it
def extract_amount(value):
    match = re.search(r'(\d+\.\d+)', str(value))
    if match:
        return float(match.group(1))
    else:
        return None
```

```python
df['Amount'] = df['Amount'].apply(extract_amount)
df = df.dropna(subset=['Amount'])
df
```

```python
# in limit and deductible column, separate the alphabetical and numeric value in two separate columns



df['Amount'].fillna(0, inplace=True)
# df['Deductible'] = df['Deductible'].str.extract(r'(\d+\.\d+)').astype(float)
# df['Limit'] = df['Limit'].str.extract(r'(\d+\.\d+)').astype(float)
df['limit_type'] = df['Limit'].str.extract(r'([a-zA-Z]+)')
df['Limit_val'] = df['Limit'].str.extract(r'(\d+)')
df['Deductible_type'] = df['Deductible'].str.extract(r'([a-zA-Z]+)')
df['Deductible_val'] = df['Deductible'].str.extract(r'(\d+)')
```

```python
df
```

```python
#drop limit, deductible as it is already separated and also drop lossdate column
df=df.drop(columns=["Limit", "Deductible", "LossDate"])
df
```

```python
#convert the categorical column into numeric col
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
col_list = ["Jurisdiction", "CoverageType", "InjuryCause", "ClassCode", "ConstructionTypeDesc","ZipCode","limit_type","Deductible_type"]
for col in col_list:
    df[col]= label_encoder.fit_transform(df[col])
df
```

```python
#convert needed col into string, integer and float value. The categorical col are converted into string. This allows machine learning algorithms to treat them as categories rather than continuous. discrete numeric variables that represent counts, IDs, or categories with ordered values, are converted to integers.


columns_to_string = ['Jurisdiction', 'CoverageType', 'InjuryCause', 'ClassCode', 'ConstructionTypeDesc', 'ZipCode', 'limit_type',
                    'Deductible_type']
df[columns_to_string] = df[columns_to_string].astype(str)
columns_to_int = ['YearBuilt', 'Stories', 'Deductible_val']
df.fillna(0, inplace=True)
df[columns_to_int] = df[columns_to_int].astype('int')
df['Limit_val'] = df['Limit_val'].astype('double')
df
```

```python
column_names = ["InjuryCause", "ConstructionTypeDesc", "ZipCode"]
value_counts_dict = {}

for column in column_names:
    value_counts_dict[column] = df[column].value_counts()

print(value_counts_dict)
```
